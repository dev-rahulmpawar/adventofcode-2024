### Part 1

```python
# Load data
p1_data_d = {}
for run_type in run_types:
    p1_data_d[run_type] = load_input_lines(os.path.join(cwd, "inputs", input_file[run_type]), label=run_type)
p1_expected = 143
```

```python
# extract page rules
p1_data1_rules_d = {}
p1_data1_pages_d = {}
for run_type in run_types:
    p1_data1_rules_d[run_type] = []
    p1_data1_pages_d[run_type] = []
    
    loading_rules = True
    for item in p1_data_d[run_type]:
        item = item.strip()
        if not item:
            loading_rules = False
            continue
        if loading_rules:
            nums = item.split("|")
            p1_data1_rules_d[run_type].append((int(nums[0]), int(nums[1])))
        else:
            nums = item.split(",")
            p1_data1_pages_d[run_type].append([int(subnum) for subnum in nums])
            
    printMultiItemList(p1_data1_rules_d[run_type], label=f"Rules : {run_type}")
    printMultiItemList(p1_data1_pages_d[run_type], label=f"Pages : {run_type}")
```

```python
# filter out pages that break any rules
p1_data2_pages_d = {}
for run_type in run_types:
    p1_data2_pages_d[run_type] = []
    p1_data1_rules = p1_data1_rules_d[run_type]
    p1_data1_pages = p1_data1_pages_d[run_type]
    
    for page_order in p1_data1_pages:
        for pageIndex, page in enumerate(page_order):
            rulefailed = True
            for rule in p1_data1_rules:
                if page in rule:
                    if page == rule[0]:
                        if rule[1] in page_order:
                            if page_order.index(rule[1]) < pageIndex:
                                break
                    else:
                        if rule[0] in page_order:
                            if page_order.index(rule[0]) > pageIndex:
                                break
            else:
                rulefailed = False
            if rulefailed:
                break
        else:
            p1_data2_pages_d[run_type].append(page_order)
            
    printMultiItemList(p1_data2_pages_d[run_type], label=f"Valid Pages : {run_type}")
```

```python
# fetch middle nums of correct pages
p1_data3_pages_d = {}
for run_type in run_types:
    p1_data3_pages_d[run_type] = [item[int((len(item) - 1)/2)] for item in p1_data2_pages_d[run_type]]
    printSingleItemList(p1_data3_pages_d[run_type], label=f"Mid Pages : {run_type}")
```

```python
# get the sum
p1_data4_d = {}
for run_type in run_types:
    p1_data4_d[run_type] = sum(p1_data3_pages_d[run_type])
    printItem(p1_data4_d[run_type], label=f"Total : {run_type}")
```

```python
# Check for sample expected
p1_data4_d["Sample"] == p1_expected
```

### Part 2

```python
# Load data
p2_new_input = False
p2_expected = 123

p2_data_d = {}
if p2_new_input:
    for run_type in run_types:
        p2_data_d[run_type] = load_input_lines(os.path.join(cwd, "inputs", input_file[run_type]), label=run_type)
else:
    p2_data_d = copy.deepcopy(p1_data_d)
    for run_type in run_types:
        printMultiItemList(p2_data_d[run_type], label=f"Input Data : {run_type}")
```

```python
# extract page rules
p2_data1_rules_d = copy.deepcopy(p1_data1_rules_d)
p2_data1_pages_d = copy.deepcopy(p1_data1_pages_d)
for run_type in run_types:
            
    printMultiItemList(p2_data1_rules_d[run_type], label=f"Rules : {run_type}")
    printMultiItemList(p2_data1_pages_d[run_type], label=f"Pages : {run_type}")
```

```python
# get pages that break any rules
p2_data2_pages_d = {}
for run_type in run_types:
    p2_data2_pages_d[run_type] =[]
    
    for page_order in p2_data1_pages_d[run_type]:
        for pageIndex, page in enumerate(page_order):
            rulefailed = True
            for rule in p2_data1_rules_d[run_type]:
                if page in rule:
                    if page == rule[0]:
                        if rule[1] in page_order:
                            if page_order.index(rule[1]) < pageIndex:
                                p2_data2_pages_d[run_type].append(page_order)
                                break
                    else:
                        if rule[0] in page_order:
                            if page_order.index(rule[0]) > pageIndex:
                                p2_data2_pages_d[run_type].append(page_order)
                                break
            else:
                rulefailed = False
            if rulefailed:
                break
    printMultiItemList(p2_data2_pages_d[run_type], label=f"Breaking Pages : {run_type}")
```

```python
def isRuleBreaking(rule, page_order, fixIt=None):
    Indx1 = page_order.index(rule[0])
    Indx2 = page_order.index(rule[1])
    breaking = Indx1 > Indx2
    if breaking and fixIt:
        page_order[Indx1], page_order[Indx2] = page_order[Indx2], page_order[Indx1]
    return breaking
        
```

```python
# fix page order
p2_data3_pages_d = {}
for run_type in run_types:
    p2_data3_pages_d[run_type] = []
    
    for page_order in p2_data2_pages_d[run_type]:
        new_order = page_order.copy()
        applicable_rules = [rule for rule in p2_data1_rules_d[run_type] if rule[0] in page_order and rule[1] in page_order]
        keepChecking = True
        while keepChecking:
            keepChecking = False
            for rule in applicable_rules:
                if isRuleBreaking(rule, new_order, True):
                    keepChecking = True
                    
        p2_data3_pages_d[run_type].append(new_order)
        
    printMultiItemList(p2_data3_pages_d[run_type], label=f"New Ordered Pages : {run_type}")
```

```python
# fetch middle nums of correct pages
p2_data4_pages_d = {}
for run_type in run_types:
    p2_data4_pages_d[run_type] = [item[int((len(item) - 1)/2)] for item in p2_data3_pages_d[run_type]]
    printSingleItemList(p2_data4_pages_d[run_type], label=f"Mid Pages : {run_type}")
```

```python
# get the sum
p2_data5_d = {}
for run_type in run_types:
    p2_data5_d[run_type] = sum(p2_data4_pages_d[run_type])
    
    printItem(p2_data5_d[run_type], label=f"Total : {run_type}")
```

```python
# Check for sample expected
p2_data5_d["Sample"] == p2_expected
```
